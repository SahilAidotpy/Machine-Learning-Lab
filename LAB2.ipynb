```python

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from random import uniform, randint
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
```

```python

x = [uniform(30, 100) for _ in range(100)]
y = [randint(0, 1) for _ in range(100)]

df1 = pd.DataFrame({"Weight": x, "Obese": y})
df1.head()
```

```python

plt.scatter(df1['Weight'], df1['Obese'], marker='*', color='red')
plt.xlabel("Weight")
plt.ylabel("Obese (0/1)")
plt.title("Weight vs Obese")
plt.show()
```

```python

model1 = LogisticRegression()


scores = cross_val_score(model1, df1[['Weight']], df1['Obese'], cv=5)

print("Cross-validation scores:", scores)
print("Mean CV Score:", np.mean(scores))
```
